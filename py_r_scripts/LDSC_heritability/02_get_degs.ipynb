# Script to get top 2k/3k differentially expressed genes (hg38 build)

```python
import os
import pandas as pd
```

```python
pd.DataFrame(ensg_top_2000).to_csv('/Volumes/broad_mcl/members_dir/mmurali/projects/9p21/DEG_hg38_list/CFA/CFA_basal_vs_hg_top_2k.GeneSet', index=False, header=False)
pd.DataFrame(ensg_top_3000).to_csv('/Volumes/broad_mcl/members_dir/mmurali/projects/9p21/DEG_hg38_list/CFA/CFA_basal_vs_hg_top_3k.GeneSet', index=False, header=False)
```

```python

input_dir = '/Volumes/broad_mcl/members_dir/mmurali/projects/9p21/DEG_hg38_list/DESeq_results_hg38'  
output_dir = '/Volumes/broad_mcl/members_dir/mmurali/projects/9p21/DEG_hg38_list/'  

def process_deg_file(file_path, output_dir):
    deg_data = pd.read_csv(file_path)
    deg_data['stat'] = pd.to_numeric(deg_data['stat'], errors='coerce')
    deg_ordered = deg_data.reindex(deg_data['stat'].abs().sort_values(ascending=False).index)
    top_2000 = deg_ordered.head(2000)
    top_3000 = deg_ordered.head(3000)
    ensg_top_2000 = top_2000['ENSG'].unique()
    ensg_top_3000 = top_3000['ENSG'].unique()

    base_name = os.path.splitext(os.path.basename(file_path))[0]
    output_2k_path = os.path.join(output_dir, f'{base_name}_top_2k.GeneSet')
    output_3k_path = os.path.join(output_dir, f'{base_name}_top_3k.GeneSet')

    pd.DataFrame(ensg_top_2000).to_csv(output_2k_path, index=False, header=False)
    pd.DataFrame(ensg_top_3000).to_csv(output_3k_path, index=False, header=False)

    print(f"Processed {base_name}: saved top 2000 to {output_2k_path}, top 3000 to {output_3k_path}")


```

```python
for file_name in os.listdir(input_dir):
    if file_name.endswith('.csv'):
        file_path = os.path.join(input_dir, file_name)
        process_deg_file(file_path, output_dir)
```
